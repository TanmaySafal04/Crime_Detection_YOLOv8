```python
import cv2
import yt_dlp
from ultralytics import YOLO

# Function to download YouTube video using yt-dlp
def download_video(url, output_path='video.mp4'):
    ydl_opts = {
        'format': 'best',
        'outtmpl': output_path,
    }
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])

# Step 1: Download the YouTube video
url = 'https://www.youtube.com/watch?v=2gY_H8Qszzs'

download_video(url, 'video.mp4')
```

```python
import cv2
from ultralytics import YOLO

# Step 2: Load the custom-trained YOLO model
model = YOLO("/teamspace/studios/this_studio/datasets/runs/detect/train/weights/best.pt")

video_path = '/teamspace/studios/this_studio/Dramatic smash-and-grab robbery at west London jewellers is caught on CCTV.mp4'
# Step 3: Process the video and perform object detection
cap = cv2.VideoCapture(video_path)
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('output.mp4', fourcc, 20.0, (int(cap.get(3)), int(cap.get(4))))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Perform object detection
    results = model(frame)

    # Draw the detection results on the frame
    for result in results.xyxy[0]:
        x1, y1, x2, y2, conf, cls = result
        label = model.names[int(cls)]
        cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
        cv2.putText(frame, f'{label} {conf:.2f}', (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    # Write the frame with the detection boxes
    out.write(frame)

cap.release()
out.release()
cv2.destroyAllWindows()
```
